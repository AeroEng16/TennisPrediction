```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.io
import ast
from plotly.subplots import make_subplots
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML/TennisPrediction/fullTrainingDataBase_140724.csv')

df.dominanceRatio = df.dominanceRatio.apply(ast.literal_eval)
df.percentageToVictory = df.percentageToVictory.apply(ast.literal_eval)
df.percentageOfGamesWon = df.percentageOfGamesWon.apply(ast.literal_eval)
df[['Player1DominanceRatio','Player2DominanceRatio']] = pd.DataFrame(df.dominanceRatio.tolist(), index= df.index)
df[['Player1percentageToVictory','Player2percentageToVictory']] = pd.DataFrame(df.percentageToVictory.tolist(), index= df.index)
df[['Player1percentageOfGamesWon','Player2percentageOfGamesWon']] = pd.DataFrame(df.percentageOfGamesWon.tolist(), index= df.index)

df["winnerAge"] = df.apply(lambda x: x[x["Winner"]+"_Age"],axis=1)
df["loserAge"] = df.apply(lambda x: x["Player2_Age"] if x["Winner"] == "Player1" else x["Player1_Age"],axis=1)

df["winnerPlayhand"] = df.apply(lambda x: x[x["Winner"]+"_PlayHand"],axis=1)
df["loserPlayhand"] = df.apply(lambda x: x["Player2_PlayHand"] if x["Winner"] == "Player1" else x["Player1_PlayHand"],axis=1)

df["winnerWeight"] = df.apply(lambda x: x[x["Winner"]+"_WeightKg"],axis=1)
df["loserWeight"] = df.apply(lambda x: x["Player2_WeightKg"] if x["Winner"] == "Player1" else x["Player1_WeightKg"],axis=1)

df["winnerHeight"] = df.apply(lambda x: x[x["Winner"]+"_HeightCm"],axis=1)
df["loserHeight"] = df.apply(lambda x: x["Player2_HeightCm"] if x["Winner"] == "Player1" else x["Player1_HeightCm"],axis=1)

df.Player1_PlayHand = df.Player1_PlayHand.apply(lambda x: 1 if x == 'R' else 0 )
df.Player2_PlayHand = df.Player2_PlayHand.apply(lambda x: 1 if x == 'R' else 0 )

# 1-hot encoding for winner( 1 for player 1, 0 for player 2)

df.Winner = df.Winner.apply(lambda x: 1 if x == 'Player1' else 0 )
df.head()
```

```python
df["EloDiff"] = df["Player1_ELO"] - df["Player2_ELO"]
df["Age_Diff"] = df["Player1_Age"] - df["Player2_Age"]
```

```python
#MODIFY TO PERCENTAGE OF GAMES WON BASED ON THESE STATS AS A FRACTION OF THE WHOLE

fig = make_subplots(rows=2,cols=2,subplot_titles=("Age","Playhand","Weight","Height"))

## Age
# Winners
fig.add_trace(
    go.Histogram(
        x=df.drop_duplicates(subset=['matchID'])["winnerAge"],name=" Winner Age"),row=1,col=1
    )
# Loser
fig.add_trace(
    go.Histogram(
        x=df.drop_duplicates(subset=['matchID'])["loserAge"],name="Loser Age"),row=1,col=1
    )
## Playhand
# Winners
fig.add_trace(
    go.Histogram(
        x=df.drop_duplicates(subset=['matchID'])["winnerPlayhand"],name="Winner Playhand"
        )
    ,row=1,col=2
    )
# Losers
fig.add_trace(
    go.Histogram(
        x=df.drop_duplicates(subset=['matchID'])["loserPlayhand"],name="Loser Playhand"
        )
    ,row=1,col=2
    )
## Weight
# Winners
fig.add_trace(
    go.Histogram(
        x=df.drop_duplicates(subset=['matchID'])["winnerWeight"],name="Winner Weight"
        )
    ,row=2,col=1
    )
# Loser
fig.add_trace(
    go.Histogram(
        x=df.drop_duplicates(subset=['matchID'])["loserWeight"],name="Loser Weight"
        )
    ,row=2,col=1
    )
## Height
# Winners
fig.add_trace(
    go.Histogram(
        x=df.drop_duplicates(subset=['matchID'])["winnerHeight"],name="Winner Height"
        )
    ,row=2,col=2
    )
# Losers
fig.add_trace(
    go.Histogram(
        x=df.drop_duplicates(subset=['matchID'])["loserHeight"],name="Loser Height"
        )
    ,row=2,col=2
    )

fig.update_layout(bargap = 0.2)
fig.update_xaxes(title_text="Age (years)", row=1, col=1)
fig.update_xaxes(title_text="Playhand", row=1, col=2)
fig.update_xaxes(title_text="Weight (kg)", row=2, col=1)
fig.update_xaxes(title_text="Height (cm)", range=[150,max(df[["Player1_HeightCm","Player2_HeightCm"]].max())*1.1], row=2, col=2)

fig.show()
plotly.io.write_json(fig, "histogram.json", pretty=True)
```
